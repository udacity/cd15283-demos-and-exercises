# Solution: Representing Time-Series Data in pandas

The city bike-share dataset has 3 years of daily ride data. Your job is to load it, set up a proper time index, inspect the date coverage, and resample at different frequencies.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
DATA_PATH = "../data/bikeshare_rides.csv"
```

## Task 1: Load and Parse Dates

Load the CSV, parse the `date` column as datetime, and set it as the index.

```python
def load_and_index(path):
    df = pd.read_csv(path, parse_dates=["date"])
    df = df.set_index("date")
    return df

df = load_and_index(DATA_PATH)
print(f"Rows: {len(df)}")
print(f"Index type: {type(df.index)}")
print(f"Date range: {df.index.min().date()} to {df.index.max().date()}")
df.head()
```

## Task 2: Check Date Coverage

Build a complete daily date range from the first to last date. Compare it to the actual index to find any missing dates.

```python
def check_coverage(df):
    full_range = pd.date_range(df.index.min(), df.index.max(), freq="D")
    return full_range.difference(df.index)

missing = check_coverage(df)
print(f"Expected days: {(df.index.max() - df.index.min()).days + 1}")
print(f"Actual days: {len(df)}")
print(f"Missing: {len(missing)}")
```

## Task 3: Resample to Weekly and Monthly

Use `.resample()` to aggregate rides to weekly (`"W"`) and monthly (`"MS"`) sums. This is how you change the frequency of a time series.

```python
def resample_series(df, freq):
    return df["rides"].resample(freq).sum()

weekly = resample_series(df, "W")
monthly = resample_series(df, "MS")
print(f"Weekly: {len(weekly)} rows")
print(f"Monthly: {len(monthly)} rows")
```

## Task 4: Plot at Different Frequencies

Plot the daily, weekly, and monthly series on separate subplots. Notice how resampling smooths out noise.

```python
def plot_frequencies(daily, weekly, monthly):
    fig, axes = plt.subplots(3, 1, figsize=(14, 10), sharex=True)
    axes[0].plot(daily.index, daily["rides"].values, linewidth=0.5)
    axes[0].set_ylabel("Rides")
    axes[0].set_title("Daily")
    axes[1].plot(weekly.index, weekly.values, linewidth=0.8)
    axes[1].set_ylabel("Rides")
    axes[1].set_title("Weekly")
    axes[2].plot(monthly.index, monthly.values, linewidth=1.0)
    axes[2].set_ylabel("Rides")
    axes[2].set_title("Monthly")
    plt.tight_layout()
    plt.show()

plot_frequencies(df, weekly, monthly)
```
