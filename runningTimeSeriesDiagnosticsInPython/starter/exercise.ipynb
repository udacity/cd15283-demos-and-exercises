Before production, determine which model is trustworthy. Fit ARIMA, SARIMAX with temperature, and a naive forecast. Run diagnostics on all three.

```
# Load and split data
df = pd.read_csv(DATA_PATH, parse_dates=["date"], index_col="date").asfreq("MS")
train = df.loc[YOUR_CODE_HERE]
test = df.loc[YOUR_CODE_HERE]
```

```
# Fit ARIMA(2,1,1) and SARIMAX(1,1,1)(1,1,1,12)+temp
arima = SARIMAX(train["demand_mwh"], order=YOUR_CODE_HERE).fit(disp=False)
sarimax = SARIMAX(train["demand_mwh"], exog=train[[YOUR_CODE_HERE]], order=YOUR_CODE_HERE, seasonal_order=YOUR_CODE_HERE).fit(disp=False)
print("Models fitted.")
```

```
# Run Ljung-Box test at lag 12 on ARIMA residuals
lb_arima = acorr_ljungbox(arima.YOUR_CODE_HERE, lags=[YOUR_CODE_HERE], return_df=True)
_, norm_p_arima = stats.normaltest(arima.YOUR_CODE_HERE.dropna())
print(f"ARIMA Ljung-Box p: {lb_arima['lb_pvalue'].values[YOUR_CODE_HERE]:.4f}")
print(f"ARIMA Normality p: {norm_p_arima:.4f}")
```

```
# Repeat for SARIMAX
lb_sarimax = acorr_ljungbox(sarimax.YOUR_CODE_HERE, lags=[YOUR_CODE_HERE], return_df=True)
_, norm_p_sarimax = stats.normaltest(sarimax.YOUR_CODE_HERE.dropna())
print(f"SARIMAX Ljung-Box p: {lb_sarimax['lb_pvalue'].values[YOUR_CODE_HERE]:.4f}")
print(f"SARIMAX Normality p: {norm_p_sarimax:.4f}")
```

```
# Build comparison table
comparison = pd.DataFrame({
    "ARIMA": [lb_arima["lb_pvalue"].values[YOUR_CODE_HERE], norm_p_arima, arima.YOUR_CODE_HERE, arima.YOUR_CODE_HERE],
    "SARIMAX": [lb_sarimax["lb_pvalue"].values[YOUR_CODE_HERE], norm_p_sarimax, sarimax.YOUR_CODE_HERE, sarimax.YOUR_CODE_HERE]
}, index=["Ljung-Box p", "Normality p", "AIC", "BIC"])
print(comparison)
```

```
fig, axes = plt.subplots(1, 2, figsize=(16, 4))
plot_acf(arima.resid, ax=axes[YOUR_CODE_HERE], lags=24, color=UB["Brand Blue"])
axes[YOUR_CODE_HERE].set_title("ARIMA Residual ACF", fontsize=18, fontweight="bold")
axes[YOUR_CODE_HERE].tick_params(axis="both", labelsize=14)
plot_acf(sarimax.resid, ax=axes[YOUR_CODE_HERE], lags=24, color=UB["Medium Blue"])
axes[YOUR_CODE_HERE].set_title("SARIMAX Residual ACF", fontsize=18, fontweight="bold")
axes[YOUR_CODE_HERE].tick_params(axis="both", labelsize=14)
plt.tight_layout()
plt.show()
```
