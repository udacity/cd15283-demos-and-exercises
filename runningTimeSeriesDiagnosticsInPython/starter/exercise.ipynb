# Exercise: Running Time-Series Diagnostics

You have two fitted models (ARIMA and SARIMAX) for electricity demand. Your job is to check whether their residuals are well-behaved and compare the models using information criteria.

- **Ljung-Box test** checks for residual autocorrelation. A high p-value means the model captured the temporal structure.
- **Normality test** validates whether prediction intervals are trustworthy.
- **AIC/BIC** let you compare models -- lower is better.

def run_residual_diagnostics(fitted):
    """
    Compute residual diagnostics for a fitted statsmodels model.
    
    Parameters
    ----------
    fitted : statsmodels results object
        A fitted SARIMAXResults object.
    
    Returns
    -------
    dict
        Keys: ljung_box_p, normality_p, aic, bic
    """
    residuals = fitted.resid
    lb = acorr_ljungbox(residuals, lags=[12], return_df=True)
    _, norm_p = stats.normaltest(residuals.dropna())
    return {
        "ljung_box_p": lb["lb_pvalue"].iloc[...],

    }

```python
def plot_residual_acf(arima_fit, sarimax_fit, lags=24):
    """
    Plot ACF of residuals for both models side by side.
    
    Parameters
    ----------
    arima_fit, sarimax_fit : statsmodels results
    lags : int
        Number of lags to display.
    """
    fig, axes = plt.subplots(1, 2, figsize=(14, 4))
    plot_acf(arima_fit.resid, ax=axes[0], lags=lags, color=...)
    axes[0].set_title("ARIMA Residuals")
    plot_acf(sarimax_fit.resid, ax=axes[1], lags=lags, color=...)
    axes[1].set_title("SARIMAX Residuals")
    for ax in axes:
        ax.tick_params(axis="both", labelsize=14)
    plt.tight_layout()
    plt.show()

plot_residual_acf(arima_fit, sarimax_fit)
```
