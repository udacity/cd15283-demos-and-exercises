## Task 4: Plot at Different Frequencies

Plot the daily, weekly, and monthly series on separate subplots. Notice how resampling smooths out noise.

```python
def plot_frequencies(daily, weekly, monthly):
    """
    Plot daily, weekly, and monthly ride counts on separate subplots.
    
    Parameters
    ----------
    daily, weekly, monthly : pd.Series
        Ride counts at each frequency.
    """
    fig, axes = plt.subplots(3, 1, figsize=(14, 12), sharex=True)
    for ax, series, title, color in zip(
        axes,
        [daily, weekly, monthly],
        ["Daily", "Weekly", "Monthly"],
        [UB["Brand Blue"], US["Orange"], US["Green"]]
    ):
        ax.plot(series.index, series.values, color=color)
        ax.set_title(title, fontsize=18, fontweight="bold")
        ax.set_ylabel("Rides", fontsize=16)
        ax.tick_params(axis="both", labelsize=14)
    plt.tight_layout()
    plt.show()

plot_frequencies(df, weekly, monthly)
```
