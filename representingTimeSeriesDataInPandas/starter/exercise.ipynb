# Exercise: Weekly Seasonality in Bike-Share Data

Load the daily bikeshare data and analyse which day of the week is busiest. Compute average rides by weekday, calculate month-over-month growth, and plot both.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import matplotlib as mpl

UB = {"Brand Blue": "#175CFF", "Medium Blue": "#7BA2FF"}
UN = {"Black": "#0A0B0F", "Dark Gray": "#5A5C62"}
mpl.rcParams["lines.linewidth"] = 3
mpl.rcParams["axes.linewidth"] = 2
```

```python
DATA_PATH = "../data/bikeshare_rides.csv"
```

```python
def load_and_clean(path):
    """
    Load bikeshare CSV, parse dates, set index, deduplicate.
    """
    df = pd.read_csv(path, parse_dates=...)
    df = df.set_index(...)
    df = df[~df.index.duplicated(keep="first")]
    return df

df = load_and_clean(DATA_PATH)
print(f"Rows: {len(df)}")
df.head()
```

```python
def weekday_averages(df):
    """
    Compute average rides for each weekday (0=Monday, 6=Sunday).
    Return a Series indexed by weekday.
    """
    return df["rides"].groupby(...).mean()

wd_avg = weekday_averages(df)
print(wd_avg)
print(f"Busiest: {wd_avg.idxmax()}")
```

```python
def monthly_growth(df):
    """
    Compute month-over-month percentage growth in total rides.
    """
    monthly = df["rides"].resample(...).sum()
    growth = monthly.pct_change() * 100
    return growth.dropna()

growth = monthly_growth(df)
print(growth.head())
```

```python
def plot_weekday_and_growth(wd_avg, growth):
    """
    Side-by-side bar chart of weekday averages and monthly growth.
    """
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
    axes[0].bar(days, wd_avg.values, color=UB["Brand Blue"])
    axes[0].set_title("Average Rides by Weekday", fontsize=18, fontweight="bold")
    axes[0].set_ylabel("Rides", fontsize=16)
    axes[0].tick_params(axis="both", labelsize=14)
    axes[1].plot(growth.index, growth.values, color=US["Orange"], marker="o")
    axes[1].axhline(y=0, color=UN["Dark Gray"], linestyle="--")
    axes[1].set_title("Month-over-Month Growth (%")", fontsize=18, fontweight="bold")
    axes[1].set_ylabel("Growth (%")", fontsize=16)
    axes[1].tick_params(axis="both", labelsize=14)
    plt.tight_layout()
    plt.show()

plot_weekday_and_growth(wd_avg, growth)
```
