## Data

`../data/bakery_sales.csv` — daily loaf sales from January 2023 through December 2024.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

HORIZON = 14  # forecast 14 days ahead
```

## Load and Explore Data

```python
df = pd.read_csv("../data/bakery_sales.csv", parse_dates=["date"], index_col="date")
sales = df["loaves_sold"]

print(f"Data: {len(sales)} days, {sales.index[0].date()} to {sales.index[-1].date()}")
print(f"Mean: {sales.mean():.0f} loaves/day")
print(f"Std:  {sales.std():.0f}")

sales.plot(figsize=(14, 4), color="black", linewidth=0.5, title="Daily Loaf Sales")
plt.ylabel("Loaves sold")
plt.tight_layout()
plt.show()
```

```python
# Weekly pattern
by_dow = sales.groupby(sales.index.dayofweek).agg(["mean", "std"])
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(days, by_dow["mean"], yerr=by_dow["std"], color="steelblue", capsize=4)
ax.set_ylabel("Loaves sold")
ax.set_title("Average Sales by Day of Week")
plt.tight_layout()
plt.show()
```
